```c++
#include <iostream>
#include <algoviz/SVG.hpp>
#include "classes/pflanze.hpp"
#include "classes/zombie.hpp"

using namespace std;

AlgoViz::clear();

SVG garten = SVG(1200, 700, "PvZ");
int killed = 0;
int anzahlSonnen = 0; 
Rect wavebar = Rect(650, 25, 200, 50, &garten);
wavebar.setFill("white");
Rect auswahlreihe[5];
Image *auswahl;
string auswahlimg[4] = {"img/erbsenkanone.png", "img/walnuss.png", "img/sonnenblume.png", "img/eiskanone.png"};
string auswahlpflanzen[4] = {"Erbsenkanone", "Walnuss", "Sonnenblume", "Eiskanone"};
int current = 0; //für die aktuelle Auswahl der Pflanze
auswahlreihe[current].setColor("red");
string auswahlzombies[5] = {"normal_zombie", "cone_zombie", "-", "-", "-"};

bool gameover = false;
int kantenlaenge = 100;
int slowtimer = 0;
int timer = 0;
int anfangstimer = 100;
int welle = 0;
int spawnrate = 5; //je weniger, desto mehr Zombies spawnen

Rect *wavehelper;
Pflanze *p;
Zombie *pz;
vector<Pflanze*> pflanzen;
vector<Zombie*> zombies;

int feld[10][5];
Rect rects[10][5];
bool besetzt[10][5];

Image bild_sun = Image("img/sonne.jpg",550,40,75,75,&garten);
Text Sonnencounter = Text("Sonnen: 0", 2 * kantenlaenge ,0.6 * kantenlaenge, &garten);
Sonnencounter.setAttribute("font-size", 50);
Image bild_zombie = Image("img/zombie.png",1170,41,70,70,&garten);
Text Killcounter = Text("Getötet: 0", 9 * kantenlaenge ,0.6 * kantenlaenge, &garten);
Killcounter.setAttribute("font-size", 50);

for (int i = 0; i < 4; i++){
    
    auswahlreihe[i].moveTo(0.5 * kantenlaenge, kantenlaenge + (i * kantenlaenge));
    auswahlreihe[i].setWidth(kantenlaenge);
    auswahlreihe[i].setHeight(0.5 * kantenlaenge);
    auswahlreihe[i].setFill("white");

    auswahlreihe[i].addTo(&garten);
    
    if (auswahlimg[i] == "img/walnuss.png") {
        auswahl = new Image (auswahlimg[i], kantenlaenge, ((i+1) * kantenlaenge) + 0.25 * kantenlaenge, 80, 80, &garten);
    } else {
        auswahl = new Image (auswahlimg[i], kantenlaenge, ((i+1) * kantenlaenge) + 0.25 * kantenlaenge, 50, 50, &garten);
    }
}

// Erstelle Feld
for (int spalte = 0 ; spalte < 10; spalte += 1) {
    for (int zeile = 0 ; zeile < 5; zeile += 1) {
        
        rects[spalte][zeile].moveTo(2 * kantenlaenge + (spalte * kantenlaenge), kantenlaenge + (zeile * kantenlaenge));
        rects[spalte][zeile].setWidth(kantenlaenge);
        rects[spalte][zeile].setHeight(kantenlaenge);
        
        if (spalte % 2 == 0) {
            if (zeile % 2 == 0) {
                rects[spalte][zeile].setFill("green");
        } else {
            rects[spalte][zeile].setFill("darkgreen");
        }
    } else {
            if (zeile % 2 == 0) {
                rects[spalte][zeile].setFill("darkgreen");
            } else {
                rects[spalte][zeile].setFill("green");
            }
        }
        rects[spalte][zeile].addTo(&garten);
        
        besetzt[spalte][zeile] = false;
    }
}

string taste = "";
MouseState maus;

AlgoViz::sendText("Töte 20 Zombies!");

vector<Pflanze*>::iterator pflanze;
vector<Zombie*>::iterator zombie;
vector<Circle*>::iterator projektil;

do {
    taste = garten.lastKey();
    maus = garten.lastClick();
    
    pflanze = pflanzen.begin();
    while (pflanze != pflanzen.end()) {
        (*pflanze)->timer += 1;
        (*pflanze)->moveProjectile(&garten); //Projektil bewegen
        
        if ((*pflanze)->name != "Sonnenblume") {
            if ((*pflanze)->timer == 50) {
                (*pflanze)->timer = 0;
                (*pflanze)->shoot(&garten); //Projektil erzeugen
            } 
            
        } else if ((*pflanze)->name == "Sonnenblume") {
            if ((*pflanze)->timer == 75) {
                    anzahlSonnen += 25;
                    Sonnencounter.setText("Sonnen: " + to_string(anzahlSonnen));
                    (*pflanze)->timer = 0;
                }
        }
        
        //Kollision
        zombie = zombies.begin();
        while (zombie != zombies.end()) {
            if ((*pflanze)->projectileList.size() != 0) {
                projektil = (*pflanze)->projectileList.begin(); //es reicht, das erste Element zu prüfen

                if ((*projektil)->getY() == (*zombie)->y && abs((*zombie)->x - (*projektil)->getX()) <= 50){
                    delete *projektil;
                    (*pflanze)->projectileList.erase(projektil);
                    (*zombie)->hit((*pflanze)->dmg);
                    if ((*pflanze)->name == "Eiskanone") {
                        (*zombie)->slow();
                    }
                } 
            }
            
            if ((*pflanze)->y == (*zombie)->y && abs((*zombie)->x - (*pflanze)->x) <= 40) {
                (*zombie)->moveSpeed = 0;
                (*zombie)->timer += 1;
                if (!(*zombie)->slowed) {
                    if ((*zombie)->timer == 50) {
                       (*pflanze)->hit((*zombie)->dmg); 
                        (*zombie)->timer = 0;
                    }
                } else {
                    if ((*zombie)->timer == 100) {
                       (*pflanze)->hit((*zombie)->dmg); 
                        (*zombie)->timer = 0;
                    }
                }
            } 

            //Zombie schneller machen wenn er geslowt wurde und die zeit dafür passt
            (*zombie)->accelerate();

            //Zombies töten
            if ((*zombie)->hp <= 0) { 
                delete *zombie; 
                zombies.erase(zombie);
                killed ++;
                Killcounter.setText("Getötet: " + to_string(killed));
            } else {
                zombie++;
            }
            if (killed == 20) {
                AlgoViz::sendText("Gewonnen!");
            }
            
        }
            //Pflanze töten
            if ((*pflanze)->hp <= 50) {
                delete *pflanze; 
                pflanzen.erase(pflanze);
                int spalte = (*pflanze)->x / 100;
                int zeile = (*pflanze)->y / 100;
                besetzt[spalte][zeile] = false;
            } else {
                pflanze++;
            }
    }
    
    
    //Zombies bewegen
    zombie = zombies.begin();
    while (zombie != zombies.end()) {
        (*zombie)->move();
        if (!(*zombie)->slowed) {
            if ((*zombie)->name == "normal_zombie") {
                (*zombie)->moveSpeed = 5;
            }
        } else {
            (*zombie)->moveSpeed = 1;
        }
        if ((*zombie)->x < 250) {
            gameover = true;
            AlgoViz::sendText("Game Over!");
        }
        zombie++;
    } 

    
    
    //Spawning Zombies:
    if(timer % (20*spawnrate)  == 0 && anfangstimer == 0){
        int zombY = (rand() % 5) + 1;  //Zombie in einer der 5 Reihen plazieren
        pz = new Zombie(auswahlzombies[0], 1200, zombY*kantenlaenge + kantenlaenge/2, &garten);
        zombies.push_back(pz);
        
        switch(welle) {
            case 1: {
                int random = rand() % 4 +1; //Random number for spawning stronger Zombies
                int compare = 1;
                if(random == compare){
                    int zombY = (rand() % 5) +1;
                    pz = new Zombie(auswahlzombies[1], 1200, zombY*kantenlaenge + kantenlaenge/2, &garten);
                    zombies.push_back(pz);
                }
                break;
            }
            case 2: {
                int random1 = (rand() & 3) +1;
                int compare1 = 2;
                if(random1 == compare1){
                    int zombY = (rand() % 5) +1;
                    pz = new Zombie(auswahlzombies[1], 1200, zombY*kantenlaenge + kantenlaenge/2, &garten);
                    zombies.push_back(pz);
                }
                break;
            }
            case 3: {
                int random2 = (rand() & 2) +1;
                int compare2 = 2;
                if(random2 == compare2){
                    int zombY = (rand() % 5) +1;
                    pz = new Zombie(auswahlzombies[1], 1200, zombY*kantenlaenge + kantenlaenge/2, &garten);
                    zombies.push_back(pz);
                }
                break; 
            }
            case 4: {
                    int zombY = (rand() % 5) +1;
                    pz = new Zombie(auswahlzombies[1], 1200, zombY*kantenlaenge + kantenlaenge/2, &garten);
                    zombies.push_back(pz);
                break;
            }
        }
    }
    
    //Sonnen aktualisieren
    if (timer % 30 == 0) {
        anzahlSonnen += 25;
        Sonnencounter.setText("Sonnen: " + to_string(anzahlSonnen));
    }

    //Welle aktualisieren
    if (timer == 300) {
        if (welle < 4) { 
            wavehelper = new Rect((welle * 50) + 650, 25, 50, 50, &garten);
            wavehelper->setFill("darkred");
            wavehelper->setColor("darkred");
            welle++;
            spawnrate--;
            timer = 0;
        }
    } 
    timer++;
    
    //Pflanzen auswählen + platzieren
    if (maus.isLegal()) {
        
        //Pflanze auswählen, wenn Maus in der Auswahlspalte 
        if (maus.x() >= auswahlreihe[0].getX() && maus.x() <= auswahlreihe[0].getX()+auswahlreihe[0].getWidth()) {
            for (int i = 0; i < 5; i++){
                auswahlreihe[i].setColor("black");
                if (maus.y() >= auswahlreihe[i].getY() && maus.y() <= auswahlreihe[i].getY()+auswahlreihe[i].getHeight()) {
                    current = i;       
                }   
            }   
            auswahlreihe[current].setColor("red"); 
        }
        
        //Pflanzen platzieren, wenn Maus im Spielfeld
        if (maus.x() >= 2 * kantenlaenge && maus.x() <= 12 * kantenlaenge) {
            if (maus.y() >= kantenlaenge && maus.y() <= 6 * kantenlaenge) {
                int spalte = maus.x() / 100;
                int zeile = maus.y() / 100;
                
                if (!besetzt[spalte][zeile]) {
                    p = new Pflanze(auswahlpflanzen[current], 10000, 10000, &garten);
                    if (anzahlSonnen >= p->cost) {
                        p = new Pflanze(auswahlpflanzen[current], (spalte * kantenlaenge) + kantenlaenge/2, (zeile * kantenlaenge) + kantenlaenge/2, &garten);
                        pflanzen.push_back(p);
                        anzahlSonnen -= p->cost;
                        Sonnencounter.setText("Sonnen: " + to_string(anzahlSonnen));
                        besetzt[spalte][zeile] = true;
                    } else {
                        delete p;
                    }
                }
            }
        }       
    }
    
    if (anfangstimer > 0) { //Zombies 
        anfangstimer--;
    }
    AlgoViz::sleep(50);
} while(!gameover && killed < 20);
```
