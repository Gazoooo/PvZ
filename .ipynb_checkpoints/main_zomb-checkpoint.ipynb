```c++
#include <iostream>
#include <algoviz/SVG.hpp>
#include "classes/erbsenkanone.hpp"
#include "classes/zombie.hpp"

using namespace std;

AlgoViz::clear();

SVG garten = SVG(1000, 500, "PvZ");
int kantenlaenge = 100;
int timer = 0;
vector<Erbsenkanone*> pflanzen;
vector<Zombie*> zombies;
Erbsenkanone *p;
Zombie *pz;
Circle *projektil;
int feld[10][5];
Rect rects[10][5];
int zTimer = 0;
// Erstelle Feld
for (int spalte = 0 ; spalte < 10; spalte += 1) {
    for (int zeile = 0 ; zeile < 5; zeile += 1) {
        
        rects[spalte][zeile].moveTo(spalte * kantenlaenge, zeile * kantenlaenge);
        rects[spalte][zeile].setWidth(kantenlaenge);
        rects[spalte][zeile].setHeight(kantenlaenge);
        
        if (spalte % 2 == 0) {
            if (zeile % 2 == 0) {
                rects[spalte][zeile].setFill("green");
        } else {
            rects[spalte][zeile].setFill("darkgreen");
        }
    } else {
            if (zeile % 2 == 0) {
                rects[spalte][zeile].setFill("darkgreen");
            } else {
                rects[spalte][zeile].setFill("green");
            }
        }
        rects[spalte][zeile].addTo(&garten);
        
    }
}

string taste = "";
MouseState maus;

AlgoViz::sendText("Beenden mit einer beliebigen Taste");

do {
    taste = garten.lastKey();
    maus = garten.lastClick();
    
    
    for (Erbsenkanone *pflanze : pflanzen) {
        pflanze->timer += 1;
        if (pflanze->timer % 800 == 0) {
            pflanze->timer = 0;
            pflanze->shoot(pflanze,&garten); //Projektil erzeugen
        } 
        if (pflanze->timer % 5 == 0) {
            //nicht so perfekt in der for schleife, da 
            //in "moveProjectile" alle Projektile bewegt werden -> 
            //es werden pro Pflanze alle Projectile bewegt
            //lsg: 5 vers. Vektoren für projectiles?
            pflanze->moveProjectile(pflanze, &garten);//Projektil bewegen
           
        }
        //Nicht funktionierende Kollision, hier wollte ich morgen weitermachen. -Arne
        //for(Zombie *zombie : zombies){
        //vector<Circle*>::iterator projektil = pflanze->projectileList.begin();
        //    while (projektil != pflanze->projectileList.end()) {
        //        if (projektil->getY() == zombie->getY() && projektil->getX() - zombie->getX() <= 50) {
        //            zombie->hit();
        //            projectileList.erase(projektil);
        //           delete *projektil;
        //        }
        //    }
        //}
    }
    
    for (Zombie *zombie : zombies) {
        if(zTimer % 1000 == 0) {
            zombie->move();
        }
        //Kollision?  Idee: Pro Reihe ein Array mit der Anzahl Klassen, für jede Klasse ein Vector in dem Array mit allen zuzeit aktiven Objekten. Problem: Könnte recht lange dauern mit dem Durchlaufen all dieser Vectoren.
        
    }
    
    if (maus.isLegal()) {
        
        int spalte = maus.x() / 100;
        int zeile = maus.y() / 100;
        
        p = new Erbsenkanone((spalte * kantenlaenge) + kantenlaenge/2, (zeile * kantenlaenge) + kantenlaenge/2, &garten);
        pflanzen.push_back(p);    
    }
    
    //Spawning Zombies:
    if(zTimer % 400 == 0){
        int zombY = (rand() % 5) +0;  //Zombie in einer der 5 Reihen plazieren
        pz = new Zombie(1000, zombY*kantenlaenge + kantenlaenge/2, &garten);
        zombies.push_back(pz);
    }
    zTimer++;
} while(taste == "");
```
